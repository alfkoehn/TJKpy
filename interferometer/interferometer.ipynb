# The interferometer at the TJ-K stellarator

## The general principle

Let's start with some general considerations. A microwave interferometer provides a non-invasive method to obtain the line-averaged electron plasma density of a plasma. The measurement is based on the principle of electromagnetic waves experiencing a phase shift when traversing a plasma compared to a wave propagating in vacuum. Measuring this phase shift then allows to determine the line-averaged electron plasma density.

To neglect the influence of the background magnetic field, we inject the probing electromagnetic wave perpendicular to it, $\vec{k}\perp\vec{B}_0$. Furthermore, we chose O-mode polarization, which is for this angle simply a linearly polarized wave with the electric field oscillating along the background magnetic field. The case thus reduces to the case without background magnetic field.

The refractive index $N$ for perpendicular injection and O-mode polarization reads
$$
N = \sqrt{1 - \left(\frac{\omega_{pe}}{\omega_0}\right)^2},
$$
with $\omega_0$ the angular frequency of the probing wave, the electron plasma frequency $\omega_{pe}$,
$$\omega_{pe}^2 = \frac{n_e e^2}{\epsilon_0 m_e},$$
with $n_e$ the electron plasma density, $e$ the elementary charge, $\epsilon_0$ the dielectric constant, and $m_e$ the mass of the electron.

To measure the electron plasma density, the probing wave is split up into two branches, one traversing the plasma and the other not, acting as the reference branch. Due to the different phase velocity in a plasma compared to vacuum, the waves have a phase difference with respect to each when comparing them again after (one of the branches) has passed the plasma. 

The phase difference $\Delta \varphi$ corresponds to a time difference:
$$\Delta \varphi 
  = \omega_0 (t_\textrm{ref} - t_\textrm{plasma}) 
  = \omega_0\left(\frac{L_\textrm{ref}}{c_0} - \frac{L_\textrm{plasma}}{v_{ph}}\right)$$,
with $c_0$ the speed of light in vacuum.

Assuming now for simplicity that $L_\textrm{ref} = L_\textrm{plasma} \equiv L$, and making use of the relation $v_{ph} = c_0/N$, we get
$$
\Delta \varphi
  = \frac{\omega_0 L}{c_0}\left( 1 - N \right)
  = \frac{\omega_0 L}{c_0}\left( 1 - \sqrt{1 - \left(\frac{\omega_{pe}}{\omega_0}\right)^2} \right).
$$

To simplify the expression, we assume that our probing frequency is far above the plasma frequency, $\omega_0\gg\omega_{pe}$, and the expression $\omega_{pe}/\omega_0$ thus very small. In the above equation, the right-hand side is partly of type $\sqrt{1-x^2}$, which we can now expand in a Taylor series around $\omega_{pe}/\omega_0\approx0$ to second order:
$$
\sqrt{1 - \left(\frac{\omega_{pe}}{\omega_0}\right)^2}
  \approx 1 - \frac{1}{2}\frac{\omega_{pe}^2}{\omega_0^2}.
$$

Inserting in the equation for the phase difference, we get
$$
\Delta \varphi 
  \approx \frac{\omega_0 L}{c_0} \left( 1 - \left(1-\frac{1}{2}\frac{\omega_{pe}^2}{\omega_0^2}\right) \right)
  = \frac{L\omega_{pe}^2}{2 \omega_0 c_0}.
$$

Inserting the expression for $\omega_{pe}$, we then get
$$
\Delta \varphi 
  = \frac{L n_e e^2}{2 \epsilon_0 m_e \omega_0 c_0}.
$$

Note that we have so far assumed that the plasma density is homogeneous, which is of course not the case. It rather has a certain profile, maybe of Gaussian or super-Gaussian shape. To account for that, we introduce the line-averaged density $\bar{n}_e$ which is defined as
$$
\bar{n}_e = \int_0^L n_e \frac{\mathrm{d}s}{L}.
$$

Inserting and rearranging, we get an expression to determine the line-averaged density from the phase difference:
$$
\bar{n}_e = \frac{2 \epsilon_0 m_e c_0}{e^2} \frac{\omega_0}{L} \Delta\varphi
$$

## Implementation of an interferometer

Let us now have a closer look at the implementation of an interferometer. As explained above, the frequency of the probing wave needs to be well above the cut-off frequency. For a typical fusion plasma, this results in probing frequencies of several $100\,\mathrm{GHz}$, even in the case of low-temperature magnetically confined plasmas, frequencies in the $100\,\mathrm{GHz}$ range should be used. Phase differences at these high frequencies are extremely hard to resolve temporarily. A solution is provided by mixing the signal down to the $\mathrm{MHz}$ regime, where commercial phase detectors are available at a reasonable price.

To mix the probing wave down, two signals with a slightly different frequency $\Delta f$ are combined in a mixer, such that the resulting signal's frequency is equal to the frequency difference of the two original signals. The high frequency part is ignored and the phase difference measured in the lower frequency part. This technique is referred to as a **heterodyne interferometer**.

### Upgrade 1 (2008)

As can be seen in the last figure, the phase detector in the original installation is not optimal: the voltage-phase relation becomes non-linear for voltages below 2 V. Therefore, Eberhard Holzhauer initiated a replacement of the phase detector, with support of the Mr. Müller from the electronic workshop. They bypassed the original phase detector with a new one based on the AD8302 which has a linear voltage-phase response. It provides, however, only a sensitive range of 180°, which is too low for some high-power shots in TJ-K. Therefore, the signal is duplicated and the duplicate shifted by 90°. These two signals are then combined/reconstructed by analog methods, extending basically the phase sensitive range to 360°. Note that the analog components used in this setup (mostly the amplifier) can slightly drift over time and the reconstruction might not be perfect and show small discontinuities at the position of reconstruction.

## Data analysis

We will now have a look at a possible implementation for calculating the line-averaged density from the measured phase difference. Python will be used here, but there are also implementation in IDL available.

We will start with importing all required libraries.

````python
# import all required libraries
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as consts
from scipy import signal
````

As a next step, we will define the probing frequency and the width of the plasma. Note that you might want to change the latter for dicharges with limiters.

````python
# define probing frequency and plasma length
f_0 = 70e9
L = 0.17
````

We will now write a simple function that translates the phase difference into line-averaged density, based on the equation derived above, where the probing frequency and the plasma length are used as input.

````python
def calc_ne(phi_in_degree, f_0=70e9, L=0.17):
    # putting all constants together
    a = 2.*consts.epsilon_0*consts.m_e*consts.c/consts.e**2
    
    # this caused some confusion: 
    # in the equation below (and derived above) phi is in radiant
    phi = phi_in_degree/180. * np.pi
    
    return f_0*2*np.pi/(2*L) * a * phi
````

As an example, we will calculate the line-averaged density for a measured phase difference of 100 degrees. Note that in the current installation, using an interferometer in the Sagnac configuration, the phase difference is measured twice and the actual phase difference (or resulting density) needs therefore to be multiplied by a factor of 0.5. See the data analysis of an actual TJ-K shot for an actual implementation of this.

````python
# current interferometer installation (Sagnac configuration)
#   ==> additional factor of 0.5 due to measuring the phase difference 2 times
delta_phi_in_degree = 100
print("f_0 = {0:.1f} GHz, L = {1:.3f} m, delta_phi = {2:.1f}".format(
    f_0*1e-9, L, delta_phi_in_degree))
print("  => n_e = {0:.3f}E17 m^-3".format(
    calc_ne(delta_phi_in_degree, f_0, L)*1e-17 * .5))
````

As a next step, we will load a timetrace of the interferometer from an actual plasma discharge done at TJ-K. Due to plasma discharges being usually refered to as 'shots' at tokamak experiments (due to their short duration), we stick to this habit and also refer plasma discharges at TJ-K as shots. Here, we will analyze shot #13277, which was performed by Izel Gediz and Alf Köhn-Seemann (Helium was used as working gas and the 2.45 GHz magnetron as heating system).

In this simplified example ("simplified" refers to the code shown here as it has no error checks and is, well, simplified), we load the column labelled ```Interferometer digital```, which corresponds to column number 9 (starting to count at 0), which is hard-coded in the following example. This is, of course, bad practise and in a proper function where we would choose the column based on its label, as this identifies the data. Check ```TJK-monitor.py``` for such an implementation.

````python
# load all timetraces acquired with tjk-monitor.vi
shot     = 13277
fname_in = 'data/shot{0:d}.dat'.format(shot)
data_all = np.loadtxt(fname_in, skiprows=4)

# extract time and interferometer data
chNr_time = 0
chNr_interf = 9
time_in_sec = data_all[:,chNr_time] * 1e-3
interf_raw = data_all[:,chNr_interf]
````

Now we are ready to make a simple plot of the interferometer signal acquired by the data acquisition. Note that this is the raw data.

````python
fig1, ax1 = plt.subplots()
ax1.plot(time_in_sec, interf_raw)

# format plot
ax1.set_title("#{0:d}".format(shot))
ax1.set_xlabel("time in seconds")
ax1.set_ylabel("signal in volts")
````

Looking at the timetrace, we can clearly see the "switch-on" of the plasma, i.e. when breakdown in the vacuum vessel happened (the jump in the signal at around 6 seconds). The end of the plasma discharge can also be clearly idendified by the jump back to the initial value of the signal (corresponding to the phase) at around 194 seconds.

Having a close look at the signal before and after plasma, we see that there is a drift in the signal which is due to thermal effects and not something happening in the plasma itself. This means we want to get rid of that drift. To do that we determine the slope between "plasma-off" in the beginning and "plasma-off" in the end and subtract that slope.


````python
# get offset at beginning and end of shot
# here, simply 100 datapoints are taken
# in a proper implementation, data until/from jump should be taken
offset_pts   = 100
offset_start = np.mean(interf_raw[0:offset_pts])
offset_end   = np.mean(interf_raw[(len(interf_raw)-offset_pts):])

# slope = delta_y / delta_x
m = (offset_end - offset_start)/(time_in_sec[-1] - time_in_sec[0])

# intersection of linear function with y-axis
b1 = offset_end - m*time_in_sec[-1]
b2 = offset_start - m*time_in_sec[0]
b  = (b1+b2)/2.

interf_noDrift = interf_raw - (m*time_in_sec + b)
````

````python
fig2, ax2 = plt.subplots()
ax2.plot(time_in_sec, interf_noDrift)

# format plot
ax2.set_title("#{0:d}, drift corrected".format(shot))
ax2.set_xlabel("time in seconds")
ax2.set_ylabel("signal in volts")
````

To smoothen the timetrace, we use a Savitzky-Golay filter, which is readily available in SciPy. Note that this smoothing might kill some of the high-frequency components you might be interested in. A more well-defined way of filtering would be a low-pass filter.

````python
interf_smooth = signal.savgol_filter(interf_noDrift, window_length=27, polyorder=1)

fig3, ax3 = plt.subplots()
ax3.plot(time_in_sec, interf_noDrift)
ax3.plot(time_in_sec, interf_smooth)

# format plot
ax3.set_title("#{0:d}, drift corrected".format(shot))
ax3.set_xlabel("time in seconds")
ax3.set_ylabel("signal in volts")
````

Plotting the raw data is only useful for debugging purposes, so we need to transform the voltage signal to the phase in degrees. This is rather easy for the AD8302, as 1 Volt corresponds to 100 degrees. Then, we can apply our above defined function which translates a phase shift into the line-averaged density. Keep in mind to use to correct frequency and plasma length.

__Note that in the current iteration of the interferometer, the phase difference is measured twice, which we need to take into account in the data analysis.__

````python
# note the factor 0.5 due to the current iteration of the interferometer
interf_density = calc_ne(interf_noDrift*100, 70e9, 0.17) * .5
interf_density_smooth = signal.savgol_filter(
    interf_density, window_length=27, polyorder=1)

fig4, ax4 = plt.subplots()
ax4.plot(time_in_sec, interf_density*1e17)
ax4.plot(time_in_sec, interf_density_smooth*1e17)

# format plot
ax4.set_title("#{0:d}, drift corrected".format(shot))
ax4.set_xlabel("time in seconds")
ax4.set_ylabel(r"$\bar{n}_e$ in $\mathrm{m}^{-3}$")
````

To get __one__ value for the line-averaged density, we can take the average of a time period at the end of the discharge. Note that one should _not_ take an average at the beginning as the operator might adjust parameters like neutral gas pressure, microwave heating power or something else. 

To define the "plasma-off" time, we look for the most negative value in the derivative of the smoothed signal.

````python
# calculate gradient of smoothed signal
interf_grad = np.gradient(interf_density_smooth)

# determine minimum of gradient to get plasma-off time
plasmaOff_id = ( np.abs(interf_grad - np.amin(interf_grad)) ).argmin()

# calculate n_e as average of a certain number of points
pts2avg = 1000
n_e = np.mean(interf_density[(plasmaOff_id-5 - pts2avg) : plasmaOff_id-5]) 
n_e_err = np.std(interf_density[(plasmaOff_id-5 - pts2avg) : plasmaOff_id-5])

print("n_e = {0:.3f} 1E17 m^-3 +- {1:.3f} 1E17 m^-3".format(n_e*1e-17, n_e_err*1e-17))
````
